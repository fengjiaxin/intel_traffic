```python
# 定义归一化函数
def normalize_data(x):
    return (x - np.mean(x))/(np.std(x))
```

```python
df_1['norm_max_value'] = df_1[['max_value']].apply(normalize_data)
```

```python
norm_data = np.array(df_1['norm_max_value'])
# 增加维度
norm_data = norm_data[:,np.newaxis]
```

```python
# 要纠正几个问题，归一化只会对特征进行归一化，之所以对max_value归一化，是因为当天的特征是之前天的label，所以有必要对label进行归一化
```
